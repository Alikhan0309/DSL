```python
import cv2
import os
```

```python
i = 0
for image in os.listdir('glasses'):
    if i > 750:
        break
    img = cv2.imread(os.path.join('glasses', image))
    print(type(img))
    if img is not None:
        # [rows, columns]
        crop = img[290:900, 210:800]
        cv2.imwrite(f"dataset/with_glasses/image-{i}.jpeg", crop)
        i += 1
    else:
        print(f"Failed to load image: {image}")
```
